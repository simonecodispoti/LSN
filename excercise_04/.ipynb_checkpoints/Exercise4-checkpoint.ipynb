## 4.1 Equilibration of phases

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# vedi foto per questa prima parte (dati a muzzum!)

# first simulation without equilibration (ne.dat)

file1 = open("jupyter_output/output_epot_ne.dat")
E_pot = [float(i) for i in file1.read().splitlines()]
file1.close()

file2 = open("jupyter_output/output_ekin_ne.dat")
E_kin = [float(i) for i in file2.read().splitlines()]
file2.close()

file3 = open("jupyter_output/output_etot_ne.dat")
E_tot = [float(i) for i in file3.read().splitlines()]
file3.close()

file4 = open("jupyter_output/output_temp_ne.dat")
T = [float(i) for i in file4.read().splitlines()]
file4.close()

TITLE_SIZE = 24
LABEL_SIZE = 16
FIG_SIZE = [8*0.5*(1+np.sqrt(5)),8]

Measures = [i for i in range(np.size(E_pot))]

fig, ((temp, kin), (pot, tot)) = plt.subplots(2, 2, constrained_layout=True, figsize=FIG_SIZE)
fig.suptitle("Instanteneous thermodynamic properties of the ensemble", fontsize=TITLE_SIZE)

temp.plot(Measures, T, color='red')
temp.set_title("Temperature", fontsize=LABEL_SIZE)
temp.set_ylabel("$T$", fontsize=LABEL_SIZE)

kin.plot(Measures, E_kin, color='orange')
kin.set_title("Kinetic energy per particle", fontsize=LABEL_SIZE)
kin.set_ylabel("$K/N$", fontsize=LABEL_SIZE)

pot.plot(Measures, E_pot, color='purple')
pot.set_title("Potential energy per particle", fontsize=LABEL_SIZE)
pot.set_ylabel("$V/N$", fontsize=LABEL_SIZE)

tot.plot(Measures, E_tot, color='blue')
tot.set_title("Total energy per particle", fontsize=LABEL_SIZE)
tot.set_ylabel("$E/N$", fontsize=LABEL_SIZE)

for i in fig.get_axes():
    i.grid(True)
    i.set_xlabel("Measurements", fontsize=LABEL_SIZE)
```

## 4.2 Average values

```python
# solid phase of Argon

from scipy import constants

Kb = constants.Boltzmann           # SI units (but energy in eV )
e = constants.elementary_charge
eps = 120*Kb                       # eps/Kb = 120 is the scale factor
eps /= e

file1 = open("jupyter_output/output_epot_ave_solid.dat")
E_pot = [eps*float(i) for i in file1.read().splitlines()]
file1.close()

file2 = open("jupyter_output/output_ekin_ave_solid.dat")
E_kin = [eps*float(i) for i in file2.read().splitlines()]
file2.close()

file3 = open("jupyter_output/output_etot_ave_solid.dat")
E_tot = [eps*float(i) for i in file3.read().splitlines()]
file3.close()

file4 = open("jupyter_output/output_temp_ave_solid.dat")
T = [120*float(i) for i in file4.read().splitlines()]
file4.close()

file5 = open("jupyter_output/output_epot_err_solid.dat")
E_pot_err = [eps*float(i) for i in file5.read().splitlines()]
file5.close()

file6 = open("jupyter_output/output_ekin_err_solid.dat")
E_kin_err = [eps*float(i) for i in file6.read().splitlines()]
file6.close()

file7 = open("jupyter_output/output_etot_err_solid.dat")
E_tot_err = [eps*float(i) for i in file7.read().splitlines()]
file7.close()

file8 = open("jupyter_output/output_temp_err_solid.dat")
T_err = [120*float(i) for i in file8.read().splitlines()]
file8.close()

n_blocks = [i for i in range(np.size(E_pot))]

fig, ((temp, kin), (pot, tot)) = plt.subplots(2, 2, constrained_layout=True, figsize=FIG_SIZE)
fig.suptitle("Average thermodynamic properties for Argon in solid phase", fontsize=TITLE_SIZE)

temp.errorbar(n_blocks, T, yerr=T_err, color='red')
temp.set_title("Temperature", fontsize=LABEL_SIZE)
temp.set_ylabel("$T \; [\, K \,]$", fontsize=LABEL_SIZE)

kin.errorbar(n_blocks, E_kin, yerr=E_kin_err, color='orange')
kin.set_title("Kinetic energy per particle", fontsize=LABEL_SIZE)
kin.set_ylabel("$K \; [\, eV \,]$", fontsize=LABEL_SIZE)

pot.errorbar(n_blocks, E_pot, yerr=E_pot_err, color='purple')
pot.set_title("Potential energy per particle", fontsize=LABEL_SIZE)
pot.set_ylabel("$V \; [\, eV \,]$", fontsize=LABEL_SIZE)

tot.errorbar(n_blocks, E_tot, yerr=E_tot_err, color='blue')
tot.set_title("Total energy per particle", fontsize=LABEL_SIZE)
tot.set_ylabel("$E \; [\, eV \,]$", fontsize=LABEL_SIZE)

for i in fig.get_axes():
    i.grid(True)
    i.set_xlabel("experiments", fontsize=LABEL_SIZE)
```
