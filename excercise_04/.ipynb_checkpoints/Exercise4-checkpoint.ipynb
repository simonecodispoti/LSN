# LSN Exercise 4

In this numerical exercise we are going to study the **average thermodynamic properties** of system of identical particles in the **NVE ensemble**, using a deterministic integration scheme for the equations of motion.

```python
# Settings and libraries:

import numpy as np
import matplotlib.pyplot as plt
from scipy import constants

import matplotlib as mpl
mpl.rcParams['axes.titlesize'] = 18
mpl.rcParams['axes.labelsize'] = 16
mpl.rcParams['legend.fontsize'] = 14
mpl.rcParams['lines.linewidth'] = 2
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['ytick.labelsize'] = 12
mpl.rcParams['axes.spines.right'] = False
mpl.rcParams['axes.spines.top'] = False
SCALE = 8
FIG_SIZE = (SCALE*0.5*(1+np.sqrt(5)),SCALE)

# Input functions:

def Read(filename):
    with open(filename) as f:
        vector = [float(i) for i in f.read().splitlines()]
    return vector

# -----------------------------------------------------------

def Read_obs(filename):
    
    a = []
    b = []
    c = []
    
    with open(filename) as f:
        data = f.read().splitlines()
    
    for i in range(len(data)):
        data[i] = data[i].split()
        a.append(float(data[i][0]))
        b.append(float(data[i][2]))
        c.append(float(data[i][3]))

    return a, b, c

# --------------------------------------------------
```

## 4.1 The equilibration problem

Our simulation start from an initial spatial configuration (file `config.0`) and set the velocities of the ensemble by sampling from a uniform distribution and *subtracting the center of mass velocity of the system*, thus eliminating a possible drift. This procedure is followed by a *rescaling*: in the NVE ensemble temperature is not fixed, but is linked to the *average kinetic energy* $\langle K \rangle$ by the relation:

$$ \langle K \rangle = \frac{3}{2} N k_B T $$

We can force the simulation at a specific average temperature by comparing the instantaneous kinetic energy $K$ with the average one for the desidered temperature $T^{\star}$ : $K^{\star} = \frac{3}{2} N k_B T^{\star}$. In reduced units ($k_B = m = 1$) the scale factor for each velocity is:

$$ \lambda = \sqrt{\frac{K^{\star}}{K}} = \sqrt{\frac{\frac{3}{2} N k_B T^{\star}} {\frac{1}{2} m \sum_{i=1}^N \mathbf{v}_i^2}} = \sqrt{\frac{3 T^{\star}}{\frac{\sum_{i=1}^N \mathbf{v}_i^2}{N}}} = 
\sqrt{\frac{3 T^{\star}}{\langle \mathbf{v}^2 \rangle}} $$

Starting from a fcc configuration, we run a simulation for the fluid phase of Argon, setting in the `input.dat` file:

$$ \rho^\star = 0.8 \quad T^\star = 1.1 \quad r_{cut} = 2.5 $$

using $10000$ integration steps and measuring temperature and energy every $10$ steps, we obtain the following trends, saved in the `non_equi` folder:

```python
E_pot = Read("non_equi/output_epot.dat")
E_kin = Read("non_equi/output_ekin.dat")
E_tot = Read("non_equi/output_etot.dat")
T = Read("non_equi/output_temp.dat")

Measures = [i for i in range(len(E_pot))]
T_target = [1.1 for i in range(len(T))]

fig, ((temp, kin), (pot, tot)) = plt.subplots(2, 2, figsize=FIG_SIZE, constrained_layout=True)
fig.suptitle("Instanteneous thermodynamic properties of the ensemble", fontsize=22)

temp.plot(Measures, T, color='red')
temp.plot(Measures, T_target, color='black', label="target")
temp.set_title("Temperature")
temp.set_ylabel("$T$")
temp.legend()

kin.plot(Measures, E_kin, color='orange')
kin.set_title("Kinetic energy per particle")
kin.set_ylabel("$K/N$")

pot.plot(Measures, E_pot, color='purple')
pot.set_title("Potential energy per particle")
pot.set_ylabel("$V/N$")

tot.plot(Measures, E_tot, color='blue')
tot.set_title("Total energy per particle")
tot.set_ylabel("$E/N$")

for i in fig.get_axes():
    i.grid(True)
    i.set_xlabel("Measurements")
```

Before reaching equilibrium, *the system undergoes a fast equilibration transition*, because **the starting configuration is not the equilibrium one**. After a MC time of order $100$, the instantaneous properties start to fluctuate around an average value, giving the possibility to estimate thermodynamic quantities of interest. The conservation of energy is guaranteed by the *Verlet algorithm*, which is symmetric and time reversable. We can use the following strategy to simulate an equilibrated system *around* a desired temperature:

- start the simulation using an fcc configuration and random velocities;
- save the final configuration and the previous one to the files `config.final` and `config.old` respectively;
- enabling the restart option in the input-file (*boolean* `old`), read from the saved configurations and extract the scaling factor $\lambda$;
- multiply every velocity by $\lambda$ and estimate a new configuration to save in the `config.old` file;
- restart the simulation many times as needed to obtain the desidered temperature range.

Starting from the final configurations of the previous simulation, we run the routine $10$ times, testing the convergence to the desidered temperature range. Every simulation is composed by $1000$ integration steps.

```python
T = Read("equi/output_temp_liq.dat")

Measures = [i for i in range(len(T))]
T_target = [1.1 for i in range(len(T))]

SCALE = 6
FIG_SIZE = (SCALE*0.5*(1+np.sqrt(5)),SCALE)
plt.figure(constrained_layout=True, figsize=FIG_SIZE)

plt.plot(Measures, T, color='red')
plt.plot(Measures, T_target, color='black', label="target")
plt.title("Temperature")
plt.xlabel("measure steps")
plt.ylabel("$T$")
plt.legend()
plt.grid()
```

From the previous plot we see that, after $10$ simulations and velocities rescaling, the system fluctuate around the desiderd equilibrium value of temperature.

## 4.2 Average properties

To study the average thermodynamic properties we adopt the **data-blocking** technique: a `walker` vector stores all the istantaneous measurements, then accumulated in a `blck_av` vector, allowing the computation of averages in each block, togehter with errors as standard deviations of the mean of the estimations in each block; averages and errors are saved in files of the form `output_obs.ave`, where *obs* is the name of the observable, in this case the *temperature*, the *kinetic energy*, the *potential energy* and the *total energy*.

As a specific Lennard-Jones fluid we choose the Argon, characterized by the energy scale $\epsilon = 120 k_B$. The scaling factor for temperatures is $\epsilon/ k_B = 120$. We adopt $eV$ units as natural scale in the atomic framework.

## 4.2.1 Solid phase

Solid phase is characterized by: $\rho^\star = 1.1$, $T^\star = 0.8$, cut-off radius = $2.2$. Before starting the true simulation to estimate avereage quantities, we search for an equilibrium configuration in the temperature range, adopting the routine described and saving results in the `equi` folder:

```python
T = Read("equi/output_temp_sol.dat")

Measures = [i for i in range(len(T))]
T_target = [0.8 for i in range(len(T))]

plt.figure(constrained_layout=True, figsize=FIG_SIZE)
plt.plot(Measures, T, color='red')
plt.plot(Measures, T_target, color='black', label="target")
plt.title("Temperature")
plt.xlabel("measure steps")
plt.ylabel("$T$")
plt.legend()
plt.grid()
```

As in the liquid case, after $10$ rescaling the ensemble stabilises at $T^{\star}=0.8$. Using $100000$ integration steps and $100$ blocks and starting from the final configurations, we obtain the following trends of the average properties of the ensemble, as a function of the number of blocks.

```python
# Rescaling factors:

Kb = constants.Boltzmann
e = constants.elementary_charge
eps = 120*Kb
eps /= e    # elettronvolt units
# ------------------------------

n_T, T, T_err = Read_obs("sol/output_temp.ave")
T = [120*i for i in T]
T_err = [120* i for i in T_err]
n_k, ekin, ekin_err = Read_obs("sol/output_ekin.ave")
ekin = [eps*i for i in ekin]
ekin_err = [eps*i for i in ekin_err]
n_p, epot, epot_err = Read_obs("sol/output_epot.ave")
epot = [eps*i for i in epot]
epot_err = [eps*i for i in epot_err]
n_e, etot, etot_err = Read_obs("sol/output_etot.ave")
etot = [eps*i for i in etot]
etot_err = [eps*i for i in etot_err]

SCALE = 8
FIG_SIZE = (SCALE*0.5*(1+np.sqrt(5)),SCALE)
fig, ((temp, kin), (pot, tot)) = plt.subplots(2, 2, constrained_layout=True, figsize=FIG_SIZE)
fig.suptitle("Average thermodynamic properties - solid phase", fontsize=22)

temp.errorbar(n_T, T, yerr=T_err, color='red')
temp.set_title("Temperature")
temp.set_ylabel("$T \; [\, K \,]$")

kin.errorbar(n_k, ekin, yerr=ekin_err, color='orange')
kin.set_title("Kinetic energy per particle")
kin.set_ylabel("$K/N \; [\, eV \,]$")

pot.errorbar(n_p, epot, yerr=epot_err, color='purple')
pot.set_title("Potential energy per particle")
pot.set_ylabel("$V/N \; [\, eV \,]$")

tot.errorbar(n_e, etot, yerr=etot_err, color='blue')
tot.set_title("Total energy per particle")
tot.set_ylabel("$E/N \; [\, eV \,]$")

for i in fig.get_axes():
    i.grid(True)
    i.set_xlabel("Number of blocks")
```

## 4.2.1 Liquid phase

For the liquid phase we use: $\rho^\star = 0.8$, $T^\star = 1.1$, cut-off radius = $2.5$. After the equilibration routine described in **section 4.1**, using $100000$ integration steps and $100$ blocks we obtain the following results:

```python
n_T, T, T_err = Read_obs("liq/output_temp.ave")
T = [120*i for i in T]
T_err = [120* i for i in T_err]
n_k, ekin, ekin_err = Read_obs("liq/output_ekin.ave")
ekin = [eps*i for i in ekin]
ekin_err = [eps*i for i in ekin_err]
n_p, epot, epot_err = Read_obs("liq/output_epot.ave")
epot = [eps*i for i in epot]
epot_err = [eps*i for i in epot_err]
n_e, etot, etot_err = Read_obs("liq/output_etot.ave")
etot = [eps*i for i in etot]
etot_err = [eps*i for i in etot_err]

fig, ((temp, kin), (pot, tot)) = plt.subplots(2, 2, constrained_layout=True, figsize=FIG_SIZE)
fig.suptitle("Average thermodynamic properties - liquid phase", fontsize=22)

temp.errorbar(n_T, T, yerr=T_err, color='red')
temp.set_title("Temperature")
temp.set_ylabel("$T \; [\, K \,]$")

kin.errorbar(n_k, ekin, yerr=ekin_err, color='orange')
kin.set_title("Kinetic energy per particle")
kin.set_ylabel("$K/N \; [\, eV \,]$")

pot.errorbar(n_p, epot, yerr=epot_err, color='purple')
pot.set_title("Potential energy per particle")
pot.set_ylabel("$V/N \; [\, eV \,]$")

tot.errorbar(n_e, etot, yerr=etot_err, color='blue')
tot.set_title("Total energy per particle")
tot.set_ylabel("$E/N \; [\, eV \,]$")

for i in fig.get_axes():
    i.grid(True)
    i.set_xlabel("Number of blocks")
```

## 4.2.1 Gas phase

To simulate the gas phase we use as inputs: $\rho^\star = 0.05$, $T^\star = 1.2$, cut-off radius = $5.0$. Again, before starting the true simulation we force the system to reach an equilibrium configuration at the temperature $T=1.2$. This phase is difficult to simulate, bevause the cut-off radius is very large. To make sure that a true equilibrium state is reached we perform many rescaling attempts, running the simulation for more steps.

```python
T = Read("equi/output_temp_gas.dat")

Measures = [i for i in range(len(T))]
T_target = [1.2 for i in range(len(T))]

SCALE = 6
FIG_SIZE = (SCALE*0.5*(1+np.sqrt(5)),SCALE)
plt.figure(constrained_layout=True, figsize=FIG_SIZE)
plt.plot(Measures, T, color='red')
plt.plot(Measures, T_target, color='black', label="target")
plt.title("Temperature")
plt.xlabel("measure steps")
plt.ylabel("$T$")
plt.legend()
plt.grid()
```

Starting from the equilibrated configuration, we use again $100000$ integration steps and $100$ blocks to estimate average quantities.

```python
n_T, T, T_err = Read_obs("gas/output_temp.ave")
T = [120*i for i in T]
T_err = [120* i for i in T_err]
n_k, ekin, ekin_err = Read_obs("gas/output_ekin.ave")
ekin = [eps*i for i in ekin]
ekin_err = [eps*i for i in ekin_err]
n_p, epot, epot_err = Read_obs("gas/output_epot.ave")
epot = [eps*i for i in epot]
epot_err = [eps*i for i in epot_err]
n_e, etot, etot_err = Read_obs("gas/output_etot.ave")
etot = [eps*i for i in etot]
etot_err = [eps*i for i in etot_err]

SCALE = 8
FIG_SIZE = (SCALE*0.5*(1+np.sqrt(5)),SCALE)
fig, ((temp, kin), (pot, tot)) = plt.subplots(2, 2, constrained_layout=True, figsize=FIG_SIZE)
fig.suptitle("Average thermodynamic properties - gas phase", fontsize=22)

temp.errorbar(n_T, T, yerr=T_err, color='red')
temp.set_title("Temperature")
temp.set_ylabel("$T \; [\, K \,]$")

kin.errorbar(n_k, ekin, yerr=ekin_err, color='orange')
kin.set_title("Kinetic energy per particle")
kin.set_ylabel("$K/N \; [\, eV \,]$")

pot.errorbar(n_p, epot, yerr=epot_err, color='purple')
pot.set_title("Potential energy per particle")
pot.set_ylabel("$V/N \; [\, eV \,]$")

tot.errorbar(n_e, etot, yerr=etot_err, color='blue')
tot.set_title("Total energy per particle")
tot.set_ylabel("$E/N \; [\, eV \,]$",)

for i in fig.get_axes():
    i.grid(True)
    i.set_xlabel("Number of blocks")
```

*In this scenario is more difficult to achieve a distinct convergence of the averages*, because the configurations vary a lot in a time interval of $10$ integration steps. To get better results we should use longer simulations, with more measurements in each block, increasing the computational effort.
