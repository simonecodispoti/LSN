## Plain vanilla option pricing

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

#si sono usate n_step = 100 estrazioni per blocco

n_cell = 100         #numero di Blocchi

file1 = open("Call_direct.txt")
call = [float(i) for i in file1.read().splitlines()]
file1.close()
    
file2 = open("Call_direct_err.txt")
call_error = [float(i) for i in file2.read().splitlines()]
file2.close()

file3 = open("Put_direct.txt")
put = [float(i) for i in file3.read().splitlines()]
file3.close()
    
file4 = open("Put_direct_err.txt")
put_error = [float(i) for i in file4.read().splitlines()]
file4.close()

N = np.arange(n_cell)

#parametri calcolati con black-scholes

call_bs = 14.975790778311286
put_bs = 5.4595325819072364

call_e = [call_bs for i in N]
put_e = [put_bs for i in N]

TITLE_SIZE = 24
LABEL_SIZE = 16

plt.figure(figsize=(6*0.5*(1+np.sqrt(5)),6), constrained_layout=True)
plt.errorbar(N, call, yerr=call_error)
plt.title("Call option price - direct", fontsize=TITLE_SIZE)
plt.xlabel("Number of simulations", fontsize=LABEL_SIZE)
plt.ylabel("$C[S(0),0]$", fontsize=LABEL_SIZE)
plt.plot(N, call_e,'r',label="expected")
plt.legend(fontsize=LABEL_SIZE)
plt.grid(True)

plt.figure(figsize=(6*0.5*(1+np.sqrt(5)),6), constrained_layout=True)
plt.errorbar(N, put, yerr=put_error)
plt.title("Put option price - direct", fontsize=TITLE_SIZE)
plt.xlabel("Number of simulations", fontsize=LABEL_SIZE)
plt.ylabel("$P[S(0),0]$", fontsize=LABEL_SIZE)
plt.plot(N, put_e,'r',label="expected")
plt.legend(fontsize=LABEL_SIZE)
plt.grid(True)
```
