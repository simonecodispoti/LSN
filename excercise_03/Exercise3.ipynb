## 3.1 Black-Scholes solution

The asset price in a fluid and ideal market evolves in time following the solution of the *Black-Scholes* equation. We choose the following *agreement parameters*:

$$ K = 100 \qquad T = 1 $$

and the following *market parameters*:

$$ S_0 = 100 \qquad r = 0.1 \qquad \sigma = 0.25 $$

integrating the equation we obtain the following estimated *call-option* and *put-option* prices.

```python
import scipy
from scipy.special import erf
from numpy.lib.scimath import sqrt, log
from numpy import exp

# Agreement parameters:
K = 100.
T = 1.

# Market parameters:
S0 = 100.
r = 0.1
sigma = 0.25

t=0

def N(x):
    return 0.5 * (1. + erf(x / sqrt(2.)))

def black_scholes(S0, K, T, r, sigma):
    d1 = 1./(sigma * sqrt(T)) * (log(S0 / K) + (r + (sigma**2) / 2.) * T)
    d2 = d1 - sigma * sqrt(T)
    C = S0 * N(d1) - K * exp(-r * T) * N(d2)
    P = S0 *(N(d1) - 1.) - K * exp(-r * T) * (N(d2)-1.)
    return C, P

# Solution:

if __name__ == "__main__":
    call_bs, put_bs = black_scholes(S0, K, T, r, sigma)
    print(" === BLACK-SCHOLES call-option and put-option pricing ===")
    print()
    print("call: ", round(call_bs,3))
    print("put:  ", round(put_bs,3))
```

## 3.2 Direct sampling method

The first method consists in sampling the final asset price at time $T$ based on the concept of going back at the time the agreement was made: based on the dsitribution of the prices $S(T)$ we want to estimate:

$$
C\,[S(0),0] = \langle e^{-rT} (S(T)-K) \rangle \qquad P\,[S(0),0] = \langle e^{-rT} (K-S(T)) \rangle
$$

where $P\,[S(0),0]$ and $C\,[S(0),0]$ denote the *put-option* and *call-option* price respectively. For fixed $T$ and $K$ we have to sample $S(T)$ from a *geometric brownian motion* with drift $r$ and volatility $\sigma$. To do this, we use our random number generator to sample from a normal distribution with mean $r$ and standard deviation $\sigma$, we evaluate $S(T)$ and we check for the exclusive conditions $S>K$ or $S<K$, collecting in two different vectors the put and the call estimantions. We do this $10000$ times and, as always, to compute averages we divide the set in $100$ blocks, with $100$ estimations each. Error bars are computed using the standard deviation of the mean. 

```python
n_cell = 100
N = np.arange(n_cell)

call = Read("Call/Call_direct.txt")
call_error = Read("Call/Call_direct_err.txt")
put = Read("Put/Put_direct.txt")
put_error = Read("Put/Put_direct_err.txt")

# Black-Scholes solution:

call_exp = [call_bs for i in N]
put_exp = [put_bs for i in N]

# Call-option:

plt.figure(figsize=FIG_SIZE, constrained_layout=True)
plt.errorbar(N, call, yerr=call_error)
plt.plot(N, call_exp, 'r', label="expected")
plt.title("Call option price - direct")
plt.xlabel("Number of simulations")
plt.ylabel("$C\,[S(0),0]$")
plt.legend()
plt.grid()

# Put-option:

plt.figure(figsize=FIG_SIZE, constrained_layout=True)
plt.errorbar(N, put, yerr=put_error)
plt.plot(N, put_exp, 'r', label="expected")
plt.title("Put option price - direct")
plt.xlabel("Number of simulations")
plt.ylabel("$P\,[S(0),0]$")
plt.ylim(4.5)
plt.legend()
plt.grid()
```

The plots above show that the Black-Scholes solution lies within the error bars of the simulated price.  

## 2.3 Discretized sampling method 

The second method is conceptually equivalent to the first one, but allows one to *simulate the entire trajectory of the geometric brownian motion*. Instead of "jumping" from $t=0$ to $t=T$, we discretize the time interval $[0,T]$ in $100$ subintervals and we proceed at steps of $dt=T/100$. Starting from $t=0$, we sample $S(dt)$, $S(2dt)$, $S(3dt)$, ... , $S(T)$ *based on the previous estimations of* $S$. This method is equivalent to the direct one because the stochastic process we simulate is a *martingale*: the expectation value of $S$ at time $t_{i+1}$ given all the values of $S$ at times $t_{i}, t_{i-1}, ... , t_{0}$ coincides with the expectation value of $S$ at time $t_{i}$. Using this strategy has a **pro** and a **con**:

-  **pro**: we can simulate the entire trajectory;
-  **con**: using the same MC steps as the direct method, we obtain the same precision for the estimation of $S(T)$ but with an increased computational effort given by the fact that a MC step is composed of other $L$ MC steps, where $L$ is the number of subintervals of $[0,T]$.

In what follows we show the sampled trajectory and the trend of $P\,[S(0),0]$ and $C\,[S(0),0]$ as a function of the number of simulated trajectories. Again we use data-blocking, with $10000$ steps == trajectories and $100$ blocks. As we expected, the results are compatible with the previous one.
