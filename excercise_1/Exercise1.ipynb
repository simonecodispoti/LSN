## 1.1 Statistical uncertainties

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

n_step = 10000       #numero di step Montecarlo
n_cell = 100         #numero di Blocchi
l = n_step/n_cell    #lunghezza dei blocchi

file1 = open("Mean.txt")
Means = file1.read().splitlines()
file1.close()

Means = [float(i) for i in Means]    
    
file2 = open("Mean_error.txt")
Errors = file2.read().splitlines()
file2.close()

Errors = [float(i) for i in Errors]

t = np.arange(n_cell)
t = t*l

expected = [0.5 for i in t]
plt.errorbar(t, Means, yerr=Errors)
plt.xlabel("Number of throws")
plt.ylabel("$<r>$")
plt.plot(t, expected,'r',label="expected $<r>=1/2$")
plt.grid(True)
plt.legend()
plt.show()
```

```python
file1 = open("Var.txt")
Means = file1.read().splitlines()
file1.close()

Means = [float(i) for i in Means]    
    
file2 = open("Var_error.txt")
Errors = file2.read().splitlines()
file2.close()

Errors = [float(i) for i in Errors]

expected = [1/12 for i in t]
plt.errorbar(t, Means, yerr=Errors)
plt.xlabel("Number of throws")
plt.ylabel("$\sigma^2$")
plt.plot(t, expected,'r',label="expected $\sigma^2 = 1/12$")
plt.grid(True)
plt.legend()
plt.show()
```
