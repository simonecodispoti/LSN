## 1.1 Statistical uncertainties

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

n_step = 10000       #numero di step Montecarlo
n_cell = 100         #numero di Blocchi
l = n_step/n_cell    #lunghezza dei blocchi

file1 = open("Mean.txt")
Means = file1.read().splitlines()
file1.close()

Means = [float(i) for i in Means]    
    
file2 = open("Mean_error.txt")
Errors = file2.read().splitlines()
file2.close()

Errors = [float(i) for i in Errors]

t = np.arange(n_cell)
t = t*l

plt.figure(figsize=(7*0.5*(1+np.sqrt(5)),7))
expected = [0.5 for i in t]
plt.errorbar(t, Means, yerr=Errors)
plt.xlabel("Number of throws")
plt.ylabel("$<r>$")
plt.plot(t, expected,'r',label="expected $<r>=1/2$")
plt.grid(True)
plt.legend()
plt.show()
```

```python
file1 = open("Var.txt")
Means = file1.read().splitlines()
file1.close()

Means = [float(i) for i in Means]    
    
file2 = open("Var_error.txt")
Errors = file2.read().splitlines()
file2.close()

Errors = [float(i) for i in Errors]

plt.figure(figsize=(7*0.5*(1+np.sqrt(5)),7))
expected = [1/12 for i in t]
plt.errorbar(t, Means, yerr=Errors)
plt.xlabel("Number of throws")
plt.ylabel("$\sigma^2$")
plt.plot(t, expected,'r',label="expected $\sigma^2 = 1/12$")
plt.grid(True)
plt.legend()
plt.show()
```

```python
file = open("Chi2.txt")
Chi2 = file.read().splitlines()
file.close()

Chi2 = [float(i) for i in Chi2]

x = [int(i) for i in range(n_cell)]

expected = [100 for i in x]

plt.figure(figsize=(7*0.5*(1+np.sqrt(5)),7))
plt.plot(x, Chi2,"ro")
plt.xlabel("Number of simulations")
plt.ylabel("$\chi^2$")
plt.plot(x, expected,'g',label="expected $\chi^2 = 100$")
plt.grid(True)
plt.legend()
plt.show()
```
